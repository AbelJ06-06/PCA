```python
!pip install matplotlib
from prody import *

from pylab import *

import matplotlib.pyplot as plt
```

```python
ion()
structure1 = parsePDB('D:/Abel/Helicobacter_pylori/PCA/control.pdb')
#pdb structure

structure2 = parsePDB("D:/Abel/Helicobacter_pylori/PCA/test.pdb")
#pdb structure

structure3 = parsePDB('D:/Abel/Helicobacter_pylori/PCA/New_C_C/control2.pdb')
#pdb structure

structure1
structure2
structure3

trajectory1 = Trajectory("D:/Abel/Helicobacter_pylori/PCA/control.dcd")
#DCD traj file
trajectory2 = Trajectory("D:/Abel/Helicobacter_pylori/PCA/test.dcd")
#DCD traj file
trajectory3 = Trajectory("D:/Abel/Helicobacter_pylori/PCA/New_C_C/control2.dcd")
#DCD traj file
```

```python
trajectory1.link(structure1)

trajectory1.setCoords(structure1)

trajectory1.setAtoms(structure1.calpha)

trajectory2.link(structure2)

trajectory2.setCoords(structure2)

trajectory2.setAtoms(structure2.calpha)

trajectory3.link(structure3)

trajectory3.setCoords(structure3)

trajectory3.setAtoms(structure3.calpha)

trajectory1
trajectory2
trajectory3
```

```python

eda1 = EDA('5fve_87203')

eda1.buildCovariance( trajectory1 )

eda1.calcModes()

eda2 = EDA('5fve_546407')

eda2.buildCovariance( trajectory2 )

eda2.calcModes()

eda3 = EDA('5fve_522832')

eda3.buildCovariance( trajectory3 )

eda3.calcModes()

eda1
eda2
```

```python
saveModel(eda1)
saveModel(eda2)
saveModel(eda3)
```

```python
# Use the 'projection' parameter to create a 3D subplot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
ax.set_xlabel('Projection on eigenvector 1')
ax.set_ylabel('Projection on eigenvector 2')
# ax.legend()

rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Times New Roman']
rcParams['font.size'] = 12

fve_87203_ca_sim = trajectory1[0:]

fve_87203_ca_sim.superpose()

fve_546407_ca_sim = trajectory2[0:]

fve_546407_ca_sim.superpose()

fve_522832_ca_sim = trajectory3[0:]

fve_522832_ca_sim.superpose()
```

```python

# We project independent trajectories in different color
showProjection(fve_87203_ca_sim, eda1[0,1], color='red', marker='.', label='5FVE_522832');

showProjection(fve_546407_ca_sim, eda2[0,1], color='blue', marker='.', label='5FVE_546407');

showProjection(fve_522832_ca_sim, eda3[0,1], color='green', marker='.', label='5FVE_87203');

# Now let's mark the beginning of the trajectory with a circle
showProjection(fve_87203_ca_sim[0], eda1[0,1], color='red', marker='o', ms=12);

showProjection(fve_546407_ca_sim[0], eda2[0,1], color='blue', marker='o', ms=12);

showProjection(fve_522832_ca_sim[0], eda3[0,1], color='green', marker='o', ms=12);

# Now let's mark the end of the trajectory with a x
showProjection(fve_87203_ca_sim[-1], eda1[0,1], color='red', marker='x', ms=12);

showProjection(fve_546407_ca_sim[-1], eda2[0,1], color='blue', marker='x', ms=12);

showProjection(fve_522832_ca_sim[-1], eda3[0,1], color='green', marker='x', ms=12);

plt.xlabel('Projection on eigenvector 2')
plt.ylabel('Projection on eigenvector 1')
plt.legend()
plt.savefig("D:\Abel\Helicobacter_pylori\PCA\output_figure_95ns_100ns_PC0_PC1.png",dpi=600)
```

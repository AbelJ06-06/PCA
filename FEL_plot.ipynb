```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

file_path = 'free-energy-landscape.dat'
# Load the data into a numpy array for processing
data = np.loadtxt(file_path, delimiter='\t')

# Extracting the x, y, and z columns
x = data[:, 0]
y = data[:, 1]
z = data[:, 2]

# Creating a grid for x and y
x_unique = np.unique(x)
y_unique = np.unique(y)

X, Y = np.meshgrid(x_unique, y_unique)

# Reshape z values to match the grid
Z = z.reshape(len(y_unique), len(x_unique))

# Plotting the surface
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Adding labels and a color bar
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('Free Energy (Z)')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

plt.title('3D Surface Plot of Free Energy Landscape')
plt.show()
```
